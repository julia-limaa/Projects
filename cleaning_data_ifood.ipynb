## 1. Organizing
At first, to better organize the df, I will drop the unnecessary columns, which are the first and the second, add the data collector's name in each df, and then rename the columns.

```python
for i in range(len(df)):
    names = ['alexandre','emanuel','emerson','julia']
    df[i] = df[i][df[i].columns[2:]] # eliminating the first and second columns
    df[i]['Dono'] = [names[i]]*len(df[i]) # adding the name

name_columns = ['T√≠tulo', 'Nota', 'Tipo', 'Tempo', 'Pre√ßo', 'Dono']

# Changing the column names in all df's
for data in df:
    data.columns = name_columns 
    
df[0].head()
```

## 2. Comma to point
Solving the comma problem in the last dataframe in the columns "Nota" and "Pre√ßo".

```python
def comma_to_point(dataframe):
    for i in ([1,4]): # appplying the function in the columns "Nota" and "Pre√ßo"
        new_column = []
        for data in dataframe.iloc[:,i]:
            if data not in (['Novo!','Gr√°tis']):
                new_data = (data.replace(',','.')) # replacing comma by period
            else:
                new_data = data
            new_column += [new_data]
        dataframe.iloc[:,i] = new_column
    return dataframe

df[3] = comma_to_point(df[3])
df[3].head()
```

# 3. Ratings
Cleaning the column "Nota".

We will see that in none of the dfs there's a rating = 0 

So I will define as 0 anything that isn't a number in this column (NAN's and "Novo!") and convert the numbers from string to float

After that, I will calculate the mean rate for each df and substitute the zeros for this mean.

```python
# Putting all the dfs in 1 dataframe
dfs_together = pd.concat(df, ignore_index = True)
dfs_together
```

```python
# Verifying that there are no rates == 0
dfs_together.loc[(dfs_together['Nota'] == 0)]
```

```python
# Function to convert numbers to float and put "0.0" where there's no number
def cleaning_ratings(dataframe):
    ratings = []
    for rate in dataframe.iloc[:,1]:
        if rate == 'Novo!':
            new_rate = float(0)
        elif pd.isnull(rate):
            new_rate = float(0)
        elif type(rate) == float:
            new_rate = rate
        elif type(rate) == str:
            new_rate = float(rate)
        ratings += [new_rate]
    dataframe.iloc[:,1] = ratings
    return dataframe
```

```python
# Applying the function to all df's
for data in df:
    data = cleaning_ratings(data)

df[1].head()
```

```python
# Substitute 0.0 to the mean of each df 
def rate_zero_to_mean(dataframe):
    mean_rate = np.mean(dataframe['Nota'])
    ratings = []
    for rate in dataframe.iloc[:,1]:
        if rate == 0:
            new_rate = round(mean_rate, 2)
        else:
            new_rate = rate
        ratings += [new_rate]
    dataframe.iloc[:,1] = ratings
    return dataframe
```

```python
# Applying the function to all df's
for data in df:
    data = rate_zero_to_mean(data)

df[1].head()
```

## 4. Food type
We will find the food type by splitting the information in the third column.

```python
def food(dataframe):
    food_types = []
    for data in dataframe.iloc[:,2]:
        splitted_data = data.split('‚Ä¢') # split the data when found a '‚Ä¢', returning a list
        if len(splitted_data) == 3:
            food_type = splitted_data[1].strip() # if lenght of the list == 3, the food type is the second item in the list
        else:
            food_type = splitted_data[0].strip() # if not, the food type is the first item
        food_types += [food_type] 
    dataframe.iloc[:,2] = food_types
    return dataframe
```

```python
# Applying the function to all df's
for data in df:
    data = food(data)

df[0].head()
```

## 5. Time 
Calculate the mean waiting time according to the range given.

```python
def r_times(dataframe):
    lst_times = []
    dataframe['Tempo'] = dataframe['Tempo'].str.replace(' min', '')
    times = dataframe['Tempo'].str.split('-')
    for time_range in times:
        sum_times = 0
        for t in time_range:
            sum_times += int(t)
        mean_time = sum_times/len(time_range)
        lst_times += [mean_time]
    dataframe.iloc[:,3] = lst_times
    return dataframe
```

```python
# Applying to all df's
for data in df:
    data = r_times(data)
    
df[1].head()
```

## 6. Price
We first need to clean up df1, which has this column different from the others, with the delivery time and delivery price

Then we will remove the "R$" from the prices.

```python
def cleaning_df1(dataframe):
    prices = []
    for data in dataframe.iloc[:,4]:
        splitted = data.split('‚Ä¢')
        price = splitted[1].strip()
        prices += [price]
    dataframe.iloc[:,4] = prices
    return dataframe
```

```python
df[0] = cleaning_df1(df[0])
df[0].head()
```

Now that the column "Pre√ßo" in the first df is similar to the others we can create a function to remove the "R$" and change 'Gr√°tis' ('Free') and NaN's to 0.0.

```python
def cleaning_prices(dataframe):
    prices = []
    for data in dataframe.iloc[:, 4]:
        if not pd.isnull(data):
            splitted = data.split('R$ ')
            if len(splitted) == 2:
                price = splitted[1].strip()
            else:
                price = splitted[0]
            if price != 'Gr√°tis':
                price = float(price)
            else:
                price = 0.0
        else:
            price = 0.0
        prices += [price]
    dataframe.iloc[:,4] = prices
    return dataframe
```

```python
for data in df:
    data = cleaning_prices(data)
    
df[0].head()
```
